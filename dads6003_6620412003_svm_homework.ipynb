```python
#HW_SVM_28/11/2567

import numpy as np


X = np.array([
    [1, 5],
    [2, 4],
    [3, 6],
    [4, 2],
    [5, 1]
])
y = np.array([1, 1, -1, -1, -1])
w = np.array([1, 0])
b = -2.5


def hinge_loss_per_datapoint(X, y, w, b):
    return np.maximum(0, 1 - y * (np.dot(X, w) + b))

losses_per_datapoint = hinge_loss_per_datapoint(X, y, w, b)

print("Hinge loss for each data point:")
for i, loss in enumerate(losses_per_datapoint):
    print(f"Data point {i+1}: {loss:.2f}")

print(f"Total hinge loss: {np.sum(losses_per_datapoint):.2f}")
```
